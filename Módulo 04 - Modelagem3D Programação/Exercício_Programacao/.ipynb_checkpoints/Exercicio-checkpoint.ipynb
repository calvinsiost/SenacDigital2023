# Exercício 2 - Gráfico de concentrações ao longo do tempo

```python
import matplotlib.pyplot as plt
from datetime import datetime

# Inserir aqui seus dados de origem
time = [datetime(2023, 1, 29), datetime(2022, 8, 3), datetime(2021, 12, 30), datetime(2020, 4, 1), datetime(2019, 10, 1), datetime(2018, 3, 1), datetime(2017, 2, 1), datetime(2016, 9, 1), datetime(2015, 3, 1), datetime(2014, 6, 9)]
concentration = [0, 1.2, 3.5, 4.2, 5.1, 6.0, 4.9, 4.3, 2.1, 1.5]
labels = ['<1', '1.2', '3.5', '4.2', '5.1', '6.0', '4.9', '4.3', '2.1', '1.5']

# Cria o gráfico cruzando tempo no eixo x e concentrações no eixo y
plt.plot(time, concentration)

# Insere os rótulos nos eixos
plt.xlabel("Tempo")
plt.ylabel("Concentrações")

# Define a concentração máxima aceitável
threshold = 5

# Cria marcadores vermelhos sempre que as concentrações forem acima do limite, e azul se forem abaixo
colors = ["red" if value >= threshold else "blue" for value in concentration]

# Insere no gráfico o esquema de cores
plt.scatter(time, concentration, c=colors)


# Adiciona a linha horizontal no eixo y com a concentração máxima aceitável
plt.axhline(y=threshold, color="red", linestyle="--")

#Título do gráfico
plt.title('Concentrações ao longo do tempo')

# Inserir os labels para cada ponto
for i in range(len(time)):
    plt.text(time[i], concentration[i], labels[i])

# Iremos exportar o gráfico como figura
plot_image_path_ex2 = 'GraficoEx2.png'
plt.savefig(plot_image_path_ex2, bbox_inches='tight')
    
# Plota o gráfico na tela
plt.show()
```

# Exercício 3 - Gráficos complexos: Diagrama Ternário

```python
import plotly.express as px
import plotly.io as pio

# Criando um dicionário para inclusão das amostras e valores das porcentagens no diagrama ternário
data = {
    "Amostras": ["Amostra S-01", "Amostra S-02", "Amostra S-03"],
    "Argila%": [0.2, 0.2, 0.5],
    "Areia%": [0.4, 0.7, 0.1],
    "Silte%": [0.4, 0.1, 0.4],
}

# Inserindo os parâmetros no gráfico
fig = px.scatter_ternary(
    data,
    a="Argila%",
    b="Areia%",
    c="Silte%",
    hover_name="Amostras",
    color="Amostras",
    title="Percentual por categoria",
)

# Plotando o gráfico na Tela
fig.show()

#Salvando a figura como imagem
plot_image_path_ex3 = 'GraficoEx3.png'
pio.write_image(fig, plot_image_path_ex3, format='png', scale=2)

# inserir uma imagem da web para comparar
from IPython.display import Image

image_url = "https://s3-sa-east-1.amazonaws.com/figuras.tecconcursos.com.br/3ecf3b58-a547-4050-9ea1-a8a181489801"
Image(url=image_url)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Carregando o arquivo csv que está salvo na mesma pasta do arquivo do Jupyter no computador / se quisermos avaliar se o dado carregou usar "print(data)"
data = pd.read_csv("data.csv")

# Atribundo as variáveis no script para cada coluna do CSV, para utilizar o algoritmo deixamos todas as variáveis númericas
## site 1 = casa de máquinas, site 2 = tanque enterrado
## DescGeology 1 = argila, DescGeology 2 = silte, DescGeology 3 = areia
X = data[["HPT", "PID", "DescGeology", "site"]]
y = data["ConcentrationWater"]

# Ajustando um modelo de regressão linear: y= B0 + B1.x + E     Onde, Y= Variável Dependente, X= Variável Independente, B0= interceptação da linha, B1 = Coeficiente de regressão linear (inclinação da linha), E = erro aleatório
## Podemos utilizar diferente modelos de machine learning, des modelos de aprenzidado supervisionado, tal como Linear Regression, Logistic Regression, Decision Trees, Random Forest, Support Vector Machines (SVM), K-Nearest Neighbors (KNN), Naive Bayes, Gradient Boosting Machines (GBM), Artificial Neural Networks (ANN). Também podemos optar por modelos de machine learning de aprendizagem não supervisionada, tal como K-Means Clustering, Hierarchical Clustering, Principal Component Analysis (PCA), Independent Component Analysis (ICA) e Non-Negative Matrix Factorization (NMF).
model = LinearRegression()
model.fit(X.values, y.values)

# Predição da variável 'ConcentrationWater' para uma nova amostra com parâmetros conhecidos
hpt_new_sample = 100
pid_new_sample = 1100
descgeology_new_sample = 2
site_new_sample = 1
new_sample = [[hpt_new_sample, pid_new_sample, descgeology_new_sample, site_new_sample]]
y_pred = model.predict(new_sample)

# Formatar resultado para 2 casas decimais e plotar na tela o valor da previsão de concentração
format_result = "{:.2f}".format( y_pred[0])
print('Predição da Concentração na Água: ' + str(format_result) + ' ug/l')

# Criar gráficos em um quadro 2x2 para cada parâmetro incluindo o valor da predição
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].scatter(data["HPT"], y, label="Amostrado")
axs[0, 0].scatter(hpt_new_sample, y_pred, label="Previsto")
axs[0, 0].set_xlabel("HPT")
axs[0, 0].set_ylabel("ConcentrationWater")
axs[0, 0].legend()

axs[0, 1].scatter(data["PID"], y, label="Amostrado")
axs[0, 1].scatter(pid_new_sample, y_pred, label="Previsto")
axs[0, 1].set_xlabel("PID")
axs[0, 1].set_ylabel("ConcentrationWater")
axs[0, 1].legend()

axs[1, 0].scatter(data["DescGeology"], y, label="Amostrado")
axs[1, 0].scatter(descgeology_new_sample, y_pred, label="Previsto")
axs[1, 0].set_xlabel("DescGeology")
axs[1, 0].set_ylabel("ConcentrationWater")
axs[1, 0].legend()

axs[1, 1].scatter(data["site"], y, label="Amostrado")
axs[1, 1].scatter(site_new_sample, y_pred, label="Previsto")
axs[1, 1].set_xlabel("site")
axs[1, 1].set_ylabel("ConcentrationWater")
axs[1, 1].legend()

# Iremos exportar o gráfico como figura
plot_image_path_ex5 = 'GraficoEx5.png'
plt.savefig(plot_image_path_ex5, bbox_inches='tight')

# Plotando o gráfico na tela
plt.show()
```
